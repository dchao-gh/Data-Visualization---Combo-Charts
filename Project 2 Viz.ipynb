```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Project_2_Data.csv')
```

```python
delta = data[data[' Brand'] == 'DELTA']
delta
```

```python
homedepot = delta[delta[' Site'] == 'Homedepot']
lowes = delta[delta[' Site'] == 'Lowes Home Improvement']
```

```python
homedepot_price = homedepot.loc[:,'PriceWeek1' : 'PriceWeek53']
homedepot_percsales = homedepot.loc[:,'PercSalesWeek1' : 'PercSalesWeek53']
homedepot_model = homedepot[' Model']
homedepot_pp = pd.concat([homedepot_model,homedepot_price], axis=1)
homedepot_final = pd.concat([homedepot_pp,homedepot_percsales], axis=1)
homedepot_final = homedepot_final.set_index(' Model')
homedepot_final
```

```python
lowes_price = lowes.loc[:,'PriceWeek1' : 'PriceWeek53']
lowes_percsales = lowes.loc[:,'PercSalesWeek1' : 'PercSalesWeek53']
lowes_model = lowes[' Model']
lowes_pp = pd.concat([lowes_model,lowes_price], axis=1)
lowes_final = pd.concat([lowes_pp,lowes_percsales], axis=1)
lowes_final = lowes_final.set_index(' Model')
lowes_final
```

```python
homedepot_pivot = homedepot_final.transpose()
lowes_pivot = lowes_final.transpose()
```

```python
dfList_hd = []
weekList = []
for i in range(1,54,1):
    weekNumber = str(i)
    weekList.append(weekNumber)
for (columnName, columnData) in homedepot_pivot.iteritems():
    x = columnData.values.tolist()
#     y = columnName
    length = len(x)
    middle_index = length // 2
    price = x[:middle_index]
    percent_sales = x[middle_index:]
    y = pd.DataFrame(price, columns = ['Price'])
    y['Percent Sales'] = percent_sales
    y['Week'] = weekList
    dfList_hd.append(y)






# for label, content in homedepot_pivot.items():
#     list_split.append(label)
#     list_split.append(content.values)
#     print(label)
#     print(content)
#     break
# type(y)
(dfList)
```

```python
dfList_low = []
weekList = []
for i in range(1,54,1):
    weekNumber = str(i)
    weekList.append(weekNumber)
for (columnName, columnData) in lowes_pivot.iteritems():
    x = columnData.values.tolist()
    length = len(x)
    middle_index = length // 2
    price = x[:middle_index]
    percent_sales = x[middle_index:]
    y = pd.DataFrame(price, columns = ['Price'])
    y['Percent Sales'] = percent_sales
    y['Week'] = weekList
    dfList_low.append(y)

dfList_low
```

```python
counter = 0
for df in dfList_hd:
    fig, ax1 = plt.subplots(figsize=(10,6))
    color = 'tab:blue'
    ax1.set_title('Price and Percentage of Sales vs Week', fontsize=16)
    ax1.set_xlabel('Week', fontsize=16)
    ax1.set_ylabel('Price', fontsize=16, color=color)
    ax2 = sns.lineplot(x='Week', y='Price', data = df)
    ax1.tick_params(axis='y')
    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Percentage of Sales', fontsize=16, color=color)
    ax2 = sns.lineplot(x='Week', y='Percent Sales', data = df, sort=False, color=color)
    ax2.tick_params(axis='y', color=color)
    counter = counter + 1
#     name = 'output.jpg'
#     number = name + str(counter)
    plt.savefig("HomeDepot{}.jpg".format(counter))
    plt.show()
#     break
```

```python
counter = 0
for df in dfList_low:
    fig, ax1 = plt.subplots(figsize=(10,6))
    color = 'tab:blue'
    ax1.set_title('Price and Percentage of Sales vs Week', fontsize=16)
    ax1.set_xlabel('Week', fontsize=16)
    ax1.set_ylabel('Price', fontsize=16, color=color)
    ax2 = sns.lineplot(x='Week', y='Price', data = df)
    ax1.tick_params(axis='y')
    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Percentage of Sales', fontsize=16, color=color)
    ax2 = sns.lineplot(x='Week', y='Percent Sales', data = df, sort=False, color=color)
    ax2.tick_params(axis='y', color=color)
    counter = counter + 1
    plt.savefig("Lowes{}.jpg".format(counter))
    plt.show()
```
